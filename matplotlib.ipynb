```python
#pyplot
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,6])
y = np.array([4,10])
plt.plot(x,y,'o')# o is just one point at the begin and end
plt.show()
```

```python
#default x ploting 0 1 2 3 4 
plt.plot(y,marker = 'o')
plt.show()
#default x ploting 0 1 2 3 4 
plt.plot(y,marker = '*')
plt.show()
#default x ploting 0 1 2 3 4 
plt.plot(y,marker = 'D')
plt.show()
```

```python
#linestyle and line color and line width
#default x ploting 0 1 2 3 4 
plt.plot(y,marker = 'D',linestyle = "dotted",color = "green",lw = "6.9")
plt.show()
#linestyle can be written as ls.

#dotted can be written as :.

#dashed can be written as --.
```

```python
#multiple lines
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
#fonts title and labels
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1,c = "green")
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
#you can set grid for x axis an
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
#plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
#So, if we want a figure with 2 rows an 1 column (meaning that the two plots will be
#displayed on top of each other instead of side-by-side), we can write the syntax like this:
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
#scatter
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)
plt.colorbar() # for the colorbar
plt.grid()
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
plt.barh(x,y,height=0.1)
plt.show()
#use twice plt.show() if u want different graphs otherwise they'll overlap
```
